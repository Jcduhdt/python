```python
#感觉requests库比urllib方便
from urllib import request
```

```python
if __name__ == "__main__":
    #以CSDN为例，CSDN不更改User Agent无法访问
    url = 'https://blog.csdn.net/c406495762/article/details/60137956'
    #需要header为字典形式
    #其实同requests一样
    head = {}
    head['User-Agent'] = 'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:66.0) Gecko/20100101 Firefox/66.0'
    #创建Request对象
    req = request.Request(url,headers=head)
    #第二种传入header方式
    #req = request.Request(url)
    #req.add_header('User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:66.0) Gecko/20100101 Firefox/66.0')
    #传入创建好的request对象
    response = request.urlopen(req)
    #读取响应信息并解码
    html = response.read().decode('utf-8')
    print(html)
```
