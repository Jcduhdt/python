```python
# -*- coding:UTF-8 -*-
#试了好多网站，只有这个网站成功了，没有编码问题，没有不显示需要的正文
#感觉是用最简单的方式爬取的，其他可能有什么反爬取机制吧，现在的我也不太懂
import requests
from bs4 import BeautifulSoup
import re
```

```python
if __name__ == "__main__":
    target = 'http://www.bxwx4.com/104027_42/2523048.html'
    req = requests.get(url = target)
    html = req.text
    bf = BeautifulSoup(html,'html.parser')#似乎在jupy上必须要加‘html.parser'进行解析
    texts = bf.find_all('div', id = 'content')
    print(texts[0].text.replace('br /','\n'))
```
